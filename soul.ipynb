```python
def initialization_D(start, matrix_for_g):
    D = []
    for i in range(len(matrix_for_g)):
        D.append(0)
    # текущие длины осбых кратчайших путей

    for i in range(len(matrix_for_g)):
        if matrix_for_g[start][i] != math.inf:
            D[i] = float(matrix_for_g[start][i])
        else:
            D[i] = float("inf")
    return D


def Dijkstra(matrix_for_g, d, start):
    S = [start]
    # множество S
    w = 0
    # текущая вершина
    P = []
    for i in range(len(matrix_for_g)):
        if start == i:
            P.append(math.inf)
        else:
            P.append(0)
    # путь в вершинах

    for i in range(len(matrix_for_g) - 1):
        fl_2 = False
        buffer_D = []
        for y in range(len(d)):
            for t in S:
                if y == t:
                    fl_2 = True
                    break
            if fl_2:
                fl_2 = False
                continue
            else:
                buffer_D.append(d[y])

        for j in range(len(d)):

            if d[j] == min(buffer_D):
                w = j
                S.append(w)
                break
        # определяем минимум

        flag = False
        for k in range(len(d)):

            for x in S:
                if k == x:
                    flag = True
                    break

            if flag:
                flag = False
                continue
            else:
                if (d[w] + matrix_for_g[w][k]) < d[k]:
                    P[k] = w

                d[k] = min(d[k], d[w] + matrix_for_g[w][k])

    return d, P


def Construct(end, p, start):
    path = [end]
    while end != start:
        path.append(p[end])
        end = p[end]
    path.reverse()

    return path




start_of_path = 0
end_of_path = 4

D_start = initialization_D(start_of_path, matrix_for_g1)

D_end, P_builder = Dijkstra(matrix_for_g1, D_start, start_of_path)

final_path = Construct(end_of_path, P_builder, start_of_path)

print(final_path)
print(D_end[end_of_path])
print(D_end)
```

```python
#BELLMANUSKA-FORDUSKA
def Construct(end, p, start):
    path = [end]
    while end != start:
        path.append(p[end])
        end = p[end]
    path.reverse()

    return path


def Inintial_D_P(start, matrix_g):
    d = [math.inf]*len(matrix_g)
    d[start] = 0
    p = [math.inf]*len(matrix_g)
    p[start] = 0

    return d, p


def Bellman_Ford(matrix_graph, d, p):
    for k in range(len(matrix_graph)):
        flag = False
        for i in range(len(matrix_graph)):
            for j in range(len(matrix_graph)):
                if d[j] > matrix_graph[i][j] + d[i]:
                    d[j] = matrix_graph[i][j] + d[i]
                    p[j] = i
                    flag = True

        if not flag:
            break

        if k == len(matrix_graph) and flag:
            print("Граф содержит цикл отрицательного веса!")

    return d, p
```

```python
#JONSON
def vertex_adder(matrix):
    for i in range(len(matrix)):
        matrix[i].append(math.inf)
    ghost_vertex = [0] * (len(matrix) + 1)
    matrix.append(ghost_vertex)

    return matrix


def Constructor_of_path(start_path, p_mega, end_path):
    path = [end_path]
    while 0 != end_path:
        end_path = p_mega[start_path][end_path]
        if end_path == 0:
            path.append(start_path)
            continue
        path.append(end_path)

    path.reverse()

    return path


def Jonson(text):
    matrix_for_graph = reader(text)

    matrix_for_graph = vertex_adder(matrix_for_graph)

    start = len(matrix_for_graph) - 1

    D, P = Inintial_D_P(start, matrix_for_graph)

    final_D, final_P = Bellman_Ford(matrix_for_graph, D, P)
    final_D.pop(-1)

    matrix_for_graph2 = reader(text)

    for i in range(len(matrix_for_graph2)):
        for j in range(len(matrix_for_graph2)):
            matrix_for_graph2[i][j] = matrix_for_graph2[i][j] + final_D[i] - final_D[j]

    MEGA_D = []
    MEGA_P = []
    for i in range(len(matrix_for_graph2)):
        start_of_path = i

        D_start = initialization_D(start_of_path, matrix_for_graph2)

        D_final, P_builder_final = Dijkstra(matrix_for_graph2, D_start, start_of_path)

        MEGA_D.append(D_final)
        MEGA_P.append(P_builder_final)

    return MEGA_D, MEGA_P
```
